```python
import pandas as pd
import json
from datetime import datetime


#Exemplo com Transferências
# 1. Criar DataFrame de clientes/contas
contas = pd.DataFrame([
    {"id_cliente": 1, "nome": "Maria", "conta": "001", "saldo": 1000.0},
    {"id_cliente": 2, "nome": "João", "conta": "002", "saldo": 500.0},
])

# 2. Criar DataFrame de transações
transacoes = pd.DataFrame(columns=["id_transacao", "conta_origem", "conta_destino", "tipo", "valor", "data"])

# Função para registrar depósitos/saques
def registrar_transacao(conta, tipo, valor):
    global contas, transacoes
    if tipo == "deposito":
        contas.loc[contas["conta"] == conta, "saldo"] += valor
    elif tipo == "saque":
        contas.loc[contas["conta"] == conta, "saldo"] -= valor
    
    nova_transacao = {
        "id_transacao": len(transacoes) + 1,
        "conta_origem": conta if tipo == "saque" else None,
        "conta_destino": conta if tipo == "deposito" else None,
        "tipo": tipo,
        "valor": valor,
        "data": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
    transacoes.loc[len(transacoes)] = nova_transacao

# Função para transferências
def transferir(conta_origem, conta_destino, valor):
    global contas, transacoes
    # Atualizar saldos
    contas.loc[contas["conta"] == conta_origem, "saldo"] -= valor
    contas.loc[contas["conta"] == conta_destino, "saldo"] += valor
    
    # Registrar transação
    nova_transacao = {
        "id_transacao": len(transacoes) + 1,
        "conta_origem": conta_origem,
        "conta_destino": conta_destino,
        "tipo": "transferencia",
        "valor": valor,
        "data": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
    transacoes.loc[len(transacoes)] = nova_transacao

# 3. Exemplos de uso
registrar_transacao("001", "deposito", 200)
registrar_transacao("002", "saque", 100)
transferir("001", "002", 150)

# 4. Relatório consolidado
relatorio = {
    "contas": contas.to_dict(orient="records"),
    "transacoes": transacoes.to_dict(orient="records")
}

# 5. Exportar para JSON
with open("relatorio_banco.json", "w", encoding="utf-8") as f:
    json.dump(relatorio, f, indent=4, ensure_ascii=False)

print("Relatório com transferências gerado com sucesso!")
```
